```python
x = datos2["celsius"]
y = datos2['fahrenheit']
```

```python
y
```

```python
x_procesada = x.values.reshape(-1,1)
y_procesado = y.values.reshape(-1,1)
print(x_procesada)
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
modelo.fit(x_procesada, y_procesado)
```

```python
modelo.predict([[43]])
```

```python
modelo.score(x_procesada, y_procesado)
```
